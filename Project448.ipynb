```python
#adding  the new feature named class
economy["class"] = "economy"
business["class"] = "business" 
```

```python
#combine the two datasets
newData = pd.concat([economy, business], ignore_index=True)
newData.head()
```

```python
newData
```

```python
# merging two columns (ch_code and "num_code") and making new feature flight
newData.num_code = newData.num_code.astype("str")
newData["flight"] = newData["ch_code"] +"-"+ newData["num_code"]
newData.drop(["ch_code","num_code"],axis = 1,inplace = True)
```

```python
newData
```

```python
# renaming the featuers
newData.rename({"dep_time": "departure_time", "from": "source_city", 
            "time_taken": "duration", "stop": "stops", "arr_time": "arrival_time",
           "to":"destination_city"}, axis = 1, inplace = True)
```

```python
newData
```

```python
new_index = range(len(newData))  # Define the new row index
newData = newData.reindex(new_index)
```

```python
newData
```

```python
print(newData['date'].unique())
```

```python
for idx, date_str in enumerate(newData['date']):
    if '/' in date_str:
        newData.at[idx, 'date'] = date_str.replace('/', '-')
```

```python
newData
```

```python
print(newData['date'].unique())
```

```python
newData['date'] = newData['date'].str.strip()

# Convert the 'date' column to datetime format
newData['date'] = pd.to_datetime(newData['date'], format='%d-%m-%Y')

# Define the reference date
reference_date = pd.to_datetime('10-02-2022', format='%d-%m-%Y')

# Calculate the difference in days and assign it to the new column 'days_left'
newData['days_left'] = (newData['date'] - reference_date).dt.days
```

```python
newData
```

```python
print(newData['date'].unique())
```

```python
print(newData['days_left'].unique())
```

```python
newData.drop("date",axis = 1,inplace = True)
```

```python
hours = pd.to_datetime(newData['departure_time']).dt.hour

# Calculate time categories
time_categories = (hours % 24 + 4) // 4
time_categories.replace({
    1: 'Late Night',
    2: 'Early Morning',
    3: 'Morning',
    4: 'Afternoon',
    5: 'Evening',
    6: 'Night'
}, inplace=True)

# Add a new column to the DataFrame with time categories
newData['departure_time'] = time_categories
```

```python
hours = pd.to_datetime(newData['arrival_time']).dt.hour

# Calculate time categories
time_categories = (hours % 24 + 4) // 4
time_categories.replace({
    1: 'Late Night',
    2: 'Early Morning',
    3: 'Morning',
    4: 'Afternoon',
    5: 'Evening',
    6: 'Night'
}, inplace=True)

# Add a new column to the DataFrame with time categories
newData['arrival_time'] = time_categories
```

```python
newData
```

```python
temp = pd.DataFrame(newData["duration"].str.split(expand=True).to_numpy().astype(str), 
                    columns=["hour", "minute"])

# Clean and convert hour column to integer
temp["hour"] = temp["hour"].apply(lambda x: re.sub("[^0-9]", "", x)).astype(int)

# Clean and replace empty strings with 0 in minute column
temp["minute"] = temp["minute"].apply(lambda r: re.sub("[^0-9]", "", r))
temp["minute"] = np.where(temp["minute"] == "", 0, temp["minute"])

# Convert minute column to integer
temp["minute"] = temp["minute"].astype(int)

# Calculate total duration in hours and round to 2 decimal places
newData["duration"] = np.around((temp["hour"] + (temp["minute"] / 60)), 2)
newData
```

```python
print(newData['stops'].unique())
```

```python
newData["stops"] = newData["stops"].apply(lambda r: re.sub("[^0-9]","",r)) # taking only digits
newData["stops"] = np.where(newData["stops"] == "", 0, newData["stops"]) # replacign "" with 0
newData["stops"] = newData["stops"].astype(int) # changing object to int
newData
```

```python
print(newData['stops'].unique())
```

```python
newData["price"] = newData["price"].apply(lambda r: re.sub("[^0-9]","",r)) # taking only digits
newData["price"] = newData["price"].astype(int) # changing object to int
newData.head(1)
```

```python
newData = newData[["airline","flight","source_city","departure_time","stops",
     "arrival_time","destination_city","class","duration","days_left","price"]]
```
