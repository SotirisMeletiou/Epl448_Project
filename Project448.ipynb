```python
newData['date'] = newData['date'].str.strip()

# Convert the 'date' column to datetime format
newData['date'] = pd.to_datetime(newData['date'], format='%d-%m-%Y')

# Define the reference date
reference_date = pd.to_datetime('10-02-2022', format='%d-%m-%Y')

# Calculate the difference in days and assign it to the new column 'days_left'
newData['days_left'] = (newData['date'] - reference_date).dt.days
```

```python
hours = pd.to_datetime(newData['departure_time']).dt.hour

# Calculate time categories
time_categories = (hours % 24 + 4) // 4
time_categories.replace({
    1: 'Late Night',
    2: 'Early Morning',
    3: 'Morning',
    4: 'Afternoon',
    5: 'Evening',
    6: 'Night'
}, inplace=True)

# Add a new column to the DataFrame with time categories
newData['departure_category'] = time_categories
```

```python
hours = pd.to_datetime(newData['arrival_time']).dt.hour

# Calculate time categories
time_categories = (hours % 24 + 4) // 4
time_categories.replace({
    1: 'Late Night',
    2: 'Early Morning',
    3: 'Morning',
    4: 'Afternoon',
    5: 'Evening',
    6: 'Night'
}, inplace=True)

# Add a new column to the DataFrame with time categories
newData['arrival_category'] = time_categories
```

```python
pattern = r'\d+\.\d+h m'
rows_with_decimal_hours = newData[newData['duration'].str.contains(pattern, na=False)]

print("Rows with durations in the format 'x.yyh m':")
print(rows_with_decimal_hours)
```

```python
corrections = {
    96486: '01h 03m',
    104676: '01h 02m',
    111315: '01h 03m',
    154007: '01h 01m'
}

# Apply corrections
for row, duration in corrections.items():
    newData.at[row, 'duration'] = duration
```

```python
def duration_to_minutes(duration_str):
    parts = duration_str.replace('h', ' ').replace('m', ' ').split()
    hours = int(parts[0]) if parts[0].isdigit() else 0
    minutes = int(parts[1]) if len(parts) > 1 and parts[1].isdigit() else 0
    total_minutes = hours * 60 + minutes
    return total_minutes

# Apply the function to the duration column
newData['duration'] = newData['duration'].apply(duration_to_minutes)
```

```python
newData["stops"] = newData["stops"].apply(lambda r: re.sub("[^0-9]","",r)) # taking only digits
newData["stops"] = np.where(newData["stops"] == "", 0, newData["stops"]) # replacign "" with 0
newData["stops"] = newData["stops"].astype(int) # changing object to int
newData
```

```python
newData["price"] = newData["price"].apply(lambda r: re.sub("[^0-9]","",r)) # taking only digits
newData["price"] = newData["price"].astype(int) # changing object to int
newData.head(1)
```

```python
def extract_hour_and_remove_zero(time_str):
    hour = int(time_str.split(':')[0])  # Extract hour part
    if hour < 10:
        return str(hour)  # Remove leading zero if hour < 10
    return time_str.split(':')[0]  # Otherwise, return the original hour string

# Apply the function to 'departure_time' and 'arrival_time' columns using .loc[]
newData.loc[:, 'departure_hour'] = newData['departure_time'].apply(extract_hour_and_remove_zero)
newData.loc[:, 'arrival_hour'] = newData['arrival_time'].apply(extract_hour_and_remove_zero)
```
